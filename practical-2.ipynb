```python
# to create a 3*3 matrix with values into a one-dimensional Numpy array

# NAME: JOSHI VIRAJ
# ENROLLMENT NO: 216270307094

import numpy as np

val = np.arange(2, 11)

mat = val.reshape(3, 3)

print(mat)
```

```python
# to append the values at the end of an array
# NAME: JOSHI VIRAJ
# ENROLLMENT NO: 216270307094

import numpy as np

arr = np.array([1, 2, 3])

print("ARRAY: ",arr)

arr2 = np.array([4, 5, 6])


appended = np.append(arr, arr2)

print("appended list: ",appended)
```

```python
# to create another shape from an array without changing its data(3*2 to 2*3)

# NAME: JOSHI VIRAJ
# ENROLLMENT NO: 216270307094

import numpy as np
original = np.array([[10, 20],
                     [30, 40],
                     [50, 60]])


reshaped = original.reshape(2, 3)

print("Original Array:")
print(original)

print("Reshaped Array:")
print(reshaped)
```
